## Workflow

```python
# Import Matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# Prepare data
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Setup plot
fig, ax = plt.subplots(figsize = (10,10)) # dimensions of the plot

# Plot data
ax.plot(x, y)

# Customize plot
ax.set(title = "Simple plot", xlabel = "x-axis", ylabel = "y-axis")

# Save figure
fig.savefig("sample_plot.png")
```

## Making figures with NumPy arrays

```python
# create some data
x = np.linspace(0, 10, 100) # creates a list of 100 evenly spaced no from 0 to 10
x[:10]
```

```python
# plot data
fig, ax = plt.subplots()
ax.plot(x,x**2);
# by default it is a line plot
```

```python
# make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# again a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# make a plot from dictionary - a bar graph
sales = {"Android": 100,
          "iOS": 200,
          "Windows": 150}
fig, ax = plt.subplots()
ax.bar(sales.keys(), sales.values());
ax.set(title = "Shop sales",
       ylabel = "No of units sold");
```

```python
fig, ax = plt.subplots()
ax.barh(list(sales.keys()), list(sales.values())); #horizontal bar, requires list
```

```python
x = np.random.randn(1000) #creates a random normal distribtion
fig, ax = plt.subplots()
ax.hist(x);
```

## Different methods for creating subplots

```python
# method 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each axes
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(sales.keys(), sales.values());
ax4.hist(np.random.randn(1000));
```

```python
# method 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# plot to different indices
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(sales.keys(), sales.values());
ax[1,1].hist(np.random.randn(1000));
```

## Plotting from pandas DataFrames

```python
# make a dataframes
carSales = pd.read_csv("data/car-sales.csv")
carSales
```

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range("1/1/2020", periods = 1000))
ts = ts.cumsum()
ts.plot()
```

```python
carSales["Price"] = carSales["Price"].str.replace('[\$\,\.]','')
carSales
```

```python
# Remove last two zeroes
carSales["Price"] = carSales["Price"].str[:-2]
carSales
```

```python
carSales["Sale Date"] = pd.date_range("1/1/2020", periods = len(carSales))
```

```python
carSales
```

```python
carSales["Total Sales"] = carSales["Price"].astype(int).cumsum()
```

```python
carSales
```

```python
#plotting the total sales
carSales.plot(x="Sale Date", y="Total Sales");
```

```python
carSales["Price"] = carSales["Price"].astype(int) # reassign price column as int
carSales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
